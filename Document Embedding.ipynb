```
from tensorflow.keras.preprocessing.text import one_hot
```

```
import numpy as np
import pandas as pd
```

```
from google.colab import files
uploaded = files.upload()

```

```
import io
df = pd.read_csv(io.BytesIO(uploaded['data_for_embeddings.txt']),error_bad_lines=False)
```

```
df
```

```
pd.get_dummies(df)
```

```
voc_size=1000000
```

```
onehot_repr=[one_hot(words,voc_size)for words in df] 
print(onehot_repr)
```

```
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```
df_length=10
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=df_length)
print(embedded_docs)
```

```
dim=25
```

```
model=Sequential()
model.add(Embedding(voc_size,10,input_length=df_length))
model.compile('adam','mse')
```

```
model.summary()
```

```
print(model.predict(embedded_docs))
```
